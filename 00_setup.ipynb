```python
import threading, time, IPython.display

def keep_alive(interval_seconds: int = 45):
    """Prevent Colab idle timeout by printing a heartbeat."""
    while True:
        time.sleep(interval_seconds)
        IPython.display.clear_output(wait=True)
        print(f"рџ’“ Session alive вЂ” {time.strftime('%Y-%m-%d %H:%M:%S')}")

t = threading.Thread(target=keep_alive, daemon=True)
t.start()
print("вњ… Keep-alive thread started (prints every 45s)")
```
